```python
#Outliers Treatment

def remove_outlier (col):
  sorted(col)
  Q1,Q3=col.quantile([0.25,0.75])
  IQR=Q3-Q1
  lower_range=Q1-(1.5*IQR)
  upper_range=Q3+(1.5*IQR)
  return  lower_range,upper_range
```

```python
lrIQ,urIQ=remove_outlier(df['IQ'])
df['IQ']=np.where(df['IQ']<lrIQ,lrIQ,df['IQ'])
df['IQ']=np.where(df['IQ']>urIQ,urIQ,df['IQ'])

df.shape
```

```python
#Scaling

from sklearn.preprocessing import StandardScaler
std_scale=StandardScaler()
std_scale
```

```python
df['College_ID']=std_scale.fit_transform(df[['College_ID']])
df['IQ']=std_scale.fit_transform(df[['IQ']])
df['Prev_Sem_Result']=std_scale.fit_transform(df[['Prev_Sem_Result']])
df['CGPA']=std_scale.fit_transform(df[['CGPA']])
df['Academic_Performance']=std_scale.fit_transform(df[['Academic_Performance']])
df['Extra_Curricular_Score']=std_scale.fit_transform(df[['Extra_Curricular_Score']])
df['Communication_Skills']=std_scale.fit_transform(df[['Communication_Skills']])
df['Projects_Completed']=std_scale.fit_transform(df[['Projects_Completed']])

df.head()
```

```python
dummies = pd.get_dummies(df[['Internship_Experience', 'Placement']], drop_first=True)
```

```python
columns=['Internship_Experience', 'Placement']
df=pd.concat([df,dummies],axis=1)

df.drop(columns,axis=1,inplace=True)
```
